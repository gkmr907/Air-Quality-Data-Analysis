```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfYear'] = df['Date'].dt.dayofyear
# Filter for Ahmedabad only
df_city= df[df['City'] == 'Ahmedabad'].copy()

# Set Date as index
df_city.set_index('Date', inplace=True)

# Display basic info
print(df_city.info())
```

```python
print(df_city.head())
```

```python
# Check for missing values
print("Missing values:\n",df_city.isnull().sum())

# Forward fill missing values for time series continuity
df_city.fillna(method='ffill', inplace=True)

# If still missing values, fill with mean for numeric columns only
numeric_cols = df_city.select_dtypes(include=np.number).columns
df_city[numeric_cols] = df_city[numeric_cols].fillna(df_city[numeric_cols].mean())

# Convert AQI to numeric (handling empty strings)
df_city['AQI'] = pd.to_numeric(df_city['AQI'], errors='coerce')

# Drop rows where AQI is still NaN (if any)
df_city.dropna(subset=['AQI'], inplace=True)

# Create a copy with only AQI for time series analysis
aqi_series = df_city[['AQI']].copy()

aqi =df_city['AQI'].dropna()
```

RATE OF CHANGE ANALYSIS

```python
# Calculate year-over-year change
annual_avg = df_city['AQI'].resample('Y').mean()
yoy_change = annual_avg.pct_change() * 100
print("Yearly AQI Change (%):")
print(yoy_change.dropna())
```

EVENT BASED ANALYSIS(DIWALI)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Diwali dates
diwali_dates = ['2015-11-11', '2016-10-30', '2017-10-19', '2018-11-07']

plt.figure(figsize=(10, 6))

for diwali_date in diwali_dates:
    # Get 7-day period centered on Diwali (3 days before to 3 days after)
    date_range = pd.date_range(
        start=pd.to_datetime(diwali_date) - pd.Timedelta(days=3),
        end=pd.to_datetime(diwali_date) + pd.Timedelta(days=3)
    )

    week_data = df_city.loc[date_range]

    #  relative day numbers (-3 to +3)
    days_from_diwali = np.arange(-3, 4)

    plt.plot(
        days_from_diwali,
        week_data['AQI'],
        marker='o',
        linestyle='-',
        linewidth=2,
        label=f'{pd.to_datetime(diwali_date).year}'
    )


plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)  # Diwali day line
plt.xticks(range(-3, 4), ['D-3', 'D-2', 'D-1', 'Diwali', 'D+1', 'D+2', 'D+3'])

plt.xlabel("Days Relative to Diwali", fontsize=12)
plt.ylabel("AQI", fontsize=12)
plt.title("AQI Levels Around Diwali (2015-2018)", fontsize=14)
plt.legend(title='Year')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

MONTHLY AVERAGES BY CITIES OVER YEARS

```python
def plot_cities_heatmap(pollutant='PM2.5', cities=None, figsize=(15, 10)):
    if cities is None:
        cities = df['City'].value_counts().head(5).index

    n_rows = len(cities)
    fig, axes = plt.subplots(n_rows, 1, figsize=figsize)

    for i, city in enumerate(cities):
        city_data = df[df['City'] == city].copy()
        city_data['Year'] = city_data['Date'].dt.year
        city_data['Month'] = city_data['Date'].dt.month

        monthly_data = city_data.pivot_table(values=pollutant,
                                           index='Year',
                                           columns='Month')

        sns.heatmap(monthly_data, cmap='YlOrRd', annot=True, fmt=".0f",
                   ax=axes[i], cbar_kws={'label': f'{pollutant} (µg/m³)'})
        axes[i].set_title(city)
        axes[i].set_xlabel('')

    plt.tight_layout()
    plt.show()

plot_cities_heatmap(cities=['Ahmedabad', 'Delhi', 'Mumbai', 'Kolkata','Chennai'])
```
